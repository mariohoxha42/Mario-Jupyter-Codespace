```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
active_data = pd.read_csv('C:/Github Repo/NBA-Analytics/data/basketballProject/prediction-files/predicted-data.csv')
cleaned_data = active_data.drop(columns=["Unnamed: 0","Unnamed: 0.1","Unnamed: 0.2","index"])
sliced_data = cleaned_data[(cleaned_data["PPG_PRED"] > 12.5) & (cleaned_data["PPG_PRED_DIFF"] < -5)]
```

```python
comp_vars = ["SEASON_NO/Season #","AGE/Age","MPG/Minutes per Game"]
for var in comp_vars:
   var_x,var_xlabel = var.split("/")
   var_y = "PPG"
   var_c = "PPG_PRED_DIFF"
   chart_title = "Points per Game as a Function of Number of Seasons"
   ax = sliced_data.plot.scatter(x=var_x, y=var_y, c=var_c, title=chart_title, s=50)

   # Annotate each data point
   for i, txt in enumerate(sliced_data["PLAYER"]):
      player_name = txt.split(" ")[-1]
      if player_name == "Jr.":
         player_name = txt.split(" ")[-2] + " " + txt.split(" ")[-1]
      ax.annotate(player_name, (sliced_data[var_x].iat[i]+0.05, sliced_data[var_y].iat[i]))

   ax.set_xlabel(var_xlabel)
   ax.set_ylabel("Points per Game")
   f = plt.gcf()
   cax = f.get_axes()[1]
   cax.set_ylabel('PPG Diff from Predicted')

   plt.show()
```
