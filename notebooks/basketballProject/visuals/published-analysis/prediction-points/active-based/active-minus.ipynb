```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
active_data = pd.read_csv('C:/Github Repo/NBA-Analytics/data/basketballProject/prediction-files/predicted-data.csv')
active_data = active_data.drop(columns=["Unnamed: 0","Unnamed: 0.1","Unnamed: 0.2","index"])
active_sliced = active_data[(active_data["PPG_PRED"] > 15) & (active_data["PPG_PRED_DIFF"] < -5)]
```

```python
multi_var_x = ["SEASON_NO","AGE","MPG"]
for var_x in multi_var_x:
   var_y = "PPG"
   var_c = "PPG_PRED_DIFF"
   chart_title = "Points per Game as a Function of Number of Seasons"
   ax = active_sliced.plot.scatter(x=var_x, y=var_y, c=var_c, title=chart_title, s=50)

   # Annotate each data point
   for i, txt in enumerate(active_sliced["PLAYER"]):
      ax.annotate(txt.split(" ")[-1], (active_sliced[var_x].iat[i]+0.05, active_sliced[var_y].iat[i]))

   ax.set_xlabel("Season #")
   ax.set_ylabel("Points per Game")
   f = plt.gcf()
   cax = f.get_axes()[1]
   cax.set_ylabel('PPG Diff from Predicted')

   plt.show()
```
