```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
historical_data = pd.read_csv('C:/Github Repo/NBA-Analytics/data/basketballProject/prediction-files/predicted-historical-data.csv')
historical_data = historical_data.drop(columns=["Unnamed: 0","Unnamed: 0.1","Unnamed: 0.2","index"])
historical_sliced = historical_data[(historical_data["PPG"] > 20) & (historical_data["PPG_PRED"] > 15) & (historical_data["PPG_PRED_DIFF"] > 10)]
```

```python
multi_var_x = ["SEASON_NO","AGE","MPG"]
for var_x in multi_var_x:
   var_y = "PPG"
   var_c = "PPG_PRED_DIFF"
   chart_title = "Points per Game as a Function of Number of Seasons"
   ax = historical_sliced.plot.scatter(x=var_x, y=var_y, c=var_c, title=chart_title, s=50)

   # Annotate each data point
   for i, txt in enumerate(historical_sliced["PLAYER"]):
      ax.annotate(txt.split(" ")[-1], (historical_sliced[var_x].iat[i]+0.05, historical_sliced[var_y].iat[i]))

   ax.set_xlabel("Season #")
   ax.set_ylabel("Points per Game")
   f = plt.gcf()
   cax = f.get_axes()[1]
   cax.set_ylabel('PPG Diff from Predicted')

   plt.show()
```
